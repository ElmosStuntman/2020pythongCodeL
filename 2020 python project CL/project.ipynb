```python
import pandas as pd
import os
import sqlite3
import matplotlib

WithDbFileDelete = True
dbfile = "excel/toyota.db"


db = sqlite3.connect(dbfile)

df1= pd.read_csv("excel/ToyotaCorolla(1).csv")

df1.to_sql("Toyota", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)


os.getcwd()
```

```python
c= db.cursor()

Create_TableName = 'cars'

Create_SQL = f"""
CREATE table if not exists {Create_TableName} as

Select ID, Model, Price, Color
From toyota;
"""

c.execute(Create_SQL)

c.close()


cars = """
select 
*
from cars;
"""

run_query(cars)
```
